Converting date columns

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

#Added new columns from order date as order by month, order by year and order by day of the week

```python
df['Order Month'] = df['Order Date'].dt.month_name()
df['Order Year'] = df['Order Date'].dt.year
df['Order Day of week'] = df['Order Date'].dt.dayofweek
df.head()
```

#Monthly sales analysis

```python

df['Order Month'] = pd.Categorical(df['Order Month'], 
                                   categories=calendar.month_name[1:], 
                                   ordered=True)
sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()


plt.figure(figsize=(10, 5))
sns.lineplot(x='Order Month', y='Sales', data=sales_by_month, marker='o')
plt.title("Total Sales by Order Month", fontsize=12)
plt.xlabel("Order Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the above plots, we conclude that the November month have highest amounts of sales while February month have lowest.

#category wise sales analysis

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()


plt.figure(figsize=(4, 5))
ax = sns.barplot(x='Category', y='Sales', data=sales_by_category)
ax.bar_label(ax.containers[0], fmt='%.0f')
plt.title("Total Sales by Category", fontsize=12)
plt.xlabel("Total Sales")
plt.ylabel("Category")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(4, 4))
plt.pie(
    sales_by_category['Sales'],
    labels=sales_by_category['Category'],
    autopct='%1.2f%%',
    startangle=90,
    colors=sns.color_palette('pastel'),
    wedgeprops={'edgecolor': 'black'}
)

plt.title("Percentage of Sales Distribution by Category", fontsize=12)
plt.tight_layout()
plt.show()
```

From the above plots it is determined that though there isn't big difference in the sales among the categories, the technology category holds highest sales while office supplies are least

Sales based on sub-categories

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index()
plt.figure(figsize=(6, 5))
ax = sns.scatterplot(x='Sub-Category', y='Sales', data=sales_by_subcategory, s=100, color='blue', marker='o')

for i in range(len(sales_by_subcategory)):
    plt.text(x=sales_by_subcategory['Sub-Category'][i],
             y=sales_by_subcategory['Sales'][i],
             s=f"{int(sales_by_subcategory['Sales'][i])}",
             ha='center',
             va='bottom',
             fontsize=6)

plt.title("Total Sales by Sub-Category", fontsize=12)
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

The above plot shows that the phones are the best sellers and Fasteners are the least sold.

#Monthly profits

```python
df['Order Month'] = pd.Categorical(df['Order Month'], 
                                   categories=calendar.month_name[1:], 
                                   ordered=True)
profits_by_month = df.groupby('Order Month')['Profit'].sum().reset_index()


plt.figure(figsize=(10, 5))
sns.lineplot(x='Order Month', y='Profit', data=profits_by_month, marker='o')
plt.title("Total Profit by Order Month", fontsize=12)
plt.xlabel("Order Month")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the plot highest profits are generated in December.

```python
profits_by_category = df.groupby('Category')['Profit'].sum().reset_index()


plt.figure(figsize=(4, 5))
ax = sns.barplot(x='Category', y='Profit', data=profits_by_category)
ax.bar_label(ax.containers[0], fmt='%.0f')
plt.title("Total Profits by Category", fontsize=12)
plt.xlabel("Total Profits")
plt.ylabel("Category")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()

plt.figure(figsize=(6, 5))
ax = sns.scatterplot(
    x='Sub-Category',
    y='Profit',
    data=profit_by_subcategory,
    s=100,
    color='blue',
    marker='o'
)

for i in range(len(profit_by_subcategory)):
    plt.text(
        x=profit_by_subcategory['Sub-Category'][i],
        y=profit_by_subcategory['Profit'][i],
        s=f"{int(profit_by_subcategory['Profit'][i])}",
        ha='center',
        va='bottom',
        fontsize=6
    )

plt.title("Total Profits by Sub-Category", fontsize=12)
plt.xlabel("Sub-Category")
plt.ylabel("Total Profits")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

From the above plot, copiers generating highest profits and tables generating lowest. 

#sales and profit by customer segment

```python
segment_data = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()


plt.figure(figsize=(4, 4))
bar_width = 0.4
x = np.arange(len(segment_data))


plt.bar(x - bar_width/2, segment_data['Sales'], width=bar_width, label='Sales', color='skyblue')

plt.bar(x + bar_width/2, segment_data['Profit'], width=bar_width, label='Profit', color='orange')

plt.xticks(x, segment_data['Segment'])
plt.ylabel("Amount")
plt.title("Sales and Profit by Customer Segment")
plt.legend()
plt.tight_layout()
plt.show()
```

#Sales to profit ratio

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

# Calculate ratio and profit margin
sales_profit_ratio = total_sales / total_profit if total_profit != 0 else np.nan
profit_margin = (total_profit / total_sales) * 100 if total_sales != 0 else np.nan

# Print the results
print(f"Total Sales: ₹{total_sales:,.2f}")
print(f"Total Profit: ₹{total_profit:,.2f}")
print(f"Sales-to-Profit Ratio: {sales_profit_ratio:.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")

# Plot sales vs profit
plt.figure(figsize=(6, 4))
bars = plt.bar(['Profit', 'Sales'], [total_profit, total_sales], color=['green', 'skyblue'])
plt.title('Total Sales vs Total Profit')
plt.ylabel('Amount (₹)')
plt.xticks(rotation=0)

# Add bar labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000, f"₹{yval:,.0f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
